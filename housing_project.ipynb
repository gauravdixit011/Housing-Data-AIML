```python
# Creating Test Set

# Testing 20%
# Training 80%
import numpy as np

def shuffle_and_split_data(data,test_ratio):
   shuffled_indices= np.random.permutation(len(data)) # yif your data size is 1500
   test_set_size = int(len(data) * test_ratio)
   test_indices = shuffled_indices[:test_set_size]  # x[:300]
   train_indices = shuffled_indices[test_set_size:] # x[300:]  
   return data.iloc[train_indices], data.iloc[test_indices]

```

```python
# 2nd Method  to split test and train set using hash id
from zlib import crc32
print(crc32(np.int64(20)))
```

```python
def is_id_in_test_set(identifier, test_ratio):
 return crc32(np.int64(identifier)) < test_ratio * 2**32

def split_data_with_id_hash(data, test_ratio, id_column):
 ids = data[id_column]
 in_test_set = ids.apply(lambda id_: is_id_in_test_set(id_, test_ratio))
 return data.loc[~in_test_set], data.loc[in_test_set]

```
